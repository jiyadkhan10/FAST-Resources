# Data analysis

```python
import pandas as pd
df=pd.read_csv("IMDB-Movie-Data.csv",index_col="Title")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df.Director.value_counts()[:10].plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title('TOP 10 DIRECTORS OF MOVIES')
```

```python
df.Actors.value_counts()[:10].plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title('TOP 10 ActorS OF MOVIES')
```

```python
df.Rating.value_counts().plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title('Rating OF MOVIES')
```

```python
#scatter plot about movie and their ratings between 2006 - 2016

plt.scatter(df.Year, df.Rating, alpha = 0.28, label = "Movie", color = "blue")
plt.xlabel("Years")
plt.ylabel("Ratings")
plt.legend(loc = "lower right")
plt.show()
```

```python
#histogram plot about number of published movies according to year

df.Year.plot(kind = "hist", bins = 40, figsize = (9,6))
plt.xlabel("Years")
plt.ylabel("Number of Movies")
plt.show()
```

```python
temp_year = df.loc[:, 'Year']
temp_year.plot(kind = 'hist', figsize=(13, 8))
#from 2006 to 2015 movie industry grows very slow, 2016 has a outbreak.
```

```python
df.plot(x='Rating', y='Revenue (Millions)',kind = 'scatter')
```

```python
df.groupby(['Title'])['Revenue (Millions)'].mean().sort_values(ascending=False)[:10].plot(kind='bar', figsize=(15,8), fontsize=13, color='green')
plt.ylabel('Revenue')
plt.title("Highest earning movies")
```

```python
df.groupby(['Title'])['Revenue (Millions)'].mean().sort_values(ascending=True)[:10].plot(kind='bar', figsize=(15,8), fontsize=13, color='green')
plt.ylabel('Revenue')
plt.title("Highest earning movies")
```

```python
#Total revenue over the years
df.groupby(['Year'])['Revenue (Millions)'].sum().plot(kind='bar', figsize=(15,8), fontsize=13, color='blue')
plt.ylabel('Revenue (Milllion USD)')
plt.title("Total Revenue By Years")
```

```python
#Total revenue over the years
df.groupby(['Year'])['Revenue (Millions)'].sum().plot(kind='line', figsize=(15,8), fontsize=13, color='blue')
plt.ylabel('Revenue (Milllion USD)')
plt.title("Total Revenue By Years")
```

```python
df.groupby(['Director'])['Revenue (Millions)'].sum().sort_values(ascending=False)[:20].plot(kind='bar', figsize=(15,8), fontsize=13, color='orange')
plt.ylabel('Revenue(in Millions USD)')
plt.title("Director-wise revenue")
```
